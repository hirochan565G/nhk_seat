```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import os
import glob

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

from scipy.signal import savgol_filter

import japanize_matplotlib
```

```python
t = np.linspace(0, 5, 300)
# print(t)
```

```python
sin5 = np.sin(2*np.pi*5*t)
sin10 = np.sin(2*np.pi*10*t)
sin25 = np.sin(2*np.pi*25*t)
sin50 = np.sin(2*np.pi*50*t)
cos100 = np.cos(2*np.pi*100*t)
mearge_5_10 = sin5 + sin10
mearge_5_25_50 = sin5 + sin25 + sin50
mearge_all = sin5 + sin10 + sin25 + sin50 + cos100
```

```python
df = pd.DataFrame({'5Hz': sin5,
                  '10Hz': sin10,
                  '25Hz': sin25,
                  '50Hz': sin50,
                  'cos100Hz': cos100,
                  '5&10Hz': mearge_5_10,
                  '5&25&50': mearge_5_25_50,
                  'all': mearge_all})
```

```python
path_now = os.getcwd()
path_project_head = os.path.dirname(path_now)
print(path_project_head)

os.mkdir(path_project_head + '/data')
path_data = os.path.join(path_project_head + '/data')
print(path_data)
```

```python
df.to_excel(path_data + '/curve.xlsx', header=True, index=None, encoding='UTF-8')
```

```python
os.listdir(path_data)
```

```python
path_data_list = glob.glob(path_data + '/*.xlsx')
path_data_str = ''.join(path_data_list)
print(path_data_str)
```
